```python
from IPython.display import HTML

HTML('''
<!DOCTYPE html>
<html lang="pt-BR">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>Rotina — Perfil Centralizado</title>
<style>
  @import url('https://fonts.googleapis.com/css2?family=Fraunces:opsz,wght@9..144,400;9..144,500;9..144,600&family=Work+Sans:wght@400;500;600&display=swap');

  * { box-sizing: border-box; }

  :root{
    --bg: #17140f;
    --card: #f3ede0;
    --card-line: #e4dbc7;
    --ink: #221e17;
    --ink-soft: #58513f;
    --muted: #948c76;
    --gold: #b8863b;
    --gold-soft: #e7cd9c;
    --teal: #3f5d56;
    --teal-soft: #d7e3df;
    --clay: #9c5738;
    --clay-soft: #ecdad0;
    --indigo: #4c5b7a;
    --indigo-soft: #dbe0ea;
    --good: #4f7a4a;
  }

  body{
    margin:0;
    background: var(--bg);
    font-family: 'Work Sans', sans-serif;
    color: var(--ink);
    padding: 40px 20px 60px;
  }

  .wrap{ max-width: 900px; margin: 0 auto; }

  .top{
    display:flex;
    align-items:center;
    gap: 28px;
    margin-bottom: 34px;
    flex-wrap: wrap;
  }

  .compass{ flex-shrink:0; }

  .headline h1{
    font-family:'Fraunces', serif;
    font-weight:600;
    font-size: 30px;
    color: #f3ede0;
    margin: 0 0 6px;
    letter-spacing: 0.2px;
  }
  .headline p{
    margin:0;
    color: #b9b09a;
    font-size: 14px;
    line-height: 1.5;
    max-width: 420px;
  }

  .legend{
    display:flex;
    gap:16px;
    margin: 0 0 28px;
    flex-wrap: wrap;
  }
  .legend span{
    display:inline-flex;
    align-items:center;
    gap:7px;
    font-size:12.5px;
    color:#c9c1ab;
  }
  .dot{ width:9px; height:9px; border-radius:50%; display:inline-block; }

  .week{
    display:grid;
    grid-template-columns: repeat(7, 1fr);
    gap: 10px;
  }

  @media (max-width: 820px){
    .week{ grid-template-columns: repeat(2, 1fr); }
  }

  .day{
    background: var(--card);
    border: 1px solid var(--card-line);
    border-radius: 14px;
    padding: 16px 14px;
    min-height: 200px;
    display:flex;
    flex-direction:column;
  }
  .day.today{ border-color: var(--gold); box-shadow: 0 0 0 1px var(--gold); }

  .day-head{
    display:flex;
    justify-content: space-between;
    align-items: baseline;
    margin-bottom: 12px;
    padding-bottom: 8px;
    border-bottom: 1px solid var(--card-line);
  }
  .day-head .dname{
    font-family:'Fraunces', serif;
    font-weight:600;
    font-size: 15px;
  }
  .day-head .ddate{
    font-size: 11px;
    color: var(--muted);
  }

  .item{
    display:flex;
    align-items:flex-start;
    gap: 9px;
    margin-bottom: 10px;
    cursor: pointer;
  }
  .item input{
    margin-top: 3px;
    width: 15px;
    height: 15px;
    accent-color: var(--gold);
    cursor:pointer;
    flex-shrink:0;
  }
  .item .cat-dot{
    width:6px; height:6px; border-radius:50%;
    margin-top: 6px;
    flex-shrink:0;
  }
  .item label{
    font-size: 12.5px;
    line-height: 1.45;
    color: var(--ink-soft);
    cursor:pointer;
  }
  .item.checked label{
    color: var(--muted);
    text-decoration: line-through;
    text-decoration-color: var(--muted);
  }

  .foot{
    margin-top: 26px;
    font-size: 12px;
    color: #857c67;
    text-align:center;
  }
</style>
</head>
<body>
<div class="wrap">

  <div class="top">
    <div class="compass">
      <svg id="ring" width="96" height="96" viewBox="0 0 96 96">
        <circle cx="48" cy="48" r="40" fill="none" stroke="#2a251d" stroke-width="7"/>
        <circle id="ring-progress" cx="48" cy="48" r="40" fill="none" stroke="#b8863b" stroke-width="7"
          stroke-linecap="round" stroke-dasharray="251.2" stroke-dashoffset="251.2"
          transform="rotate(-90 48 48)"/>
        <text id="ring-pct" x="48" y="44" text-anchor="middle" font-family="Fraunces, serif" font-size="18" font-weight="600" fill="#f3ede0">0%</text>
        <text x="48" y="60" text-anchor="middle" font-family="Work Sans, sans-serif" font-size="8.5" fill="#b9b09a">na semana</text>
      </svg>
    </div>
    <div class="headline">
      <h1>Rotina do perfil Centralizado</h1>
      <p>Práticas do relatório Recognita organizadas ao longo da semana, para manter a flexibilidade cognitiva e desenvolver autoconhecimento e concentração.</p>
    </div>
  </div>

  <div class="legend">
    <span><i class="dot" style="background:var(--indigo)"></i>Flexibilidade cognitiva</span>
    <span><i class="dot" style="background:var(--clay)"></i>Autoconhecimento</span>
    <span><i class="dot" style="background:var(--teal)"></i>Concentração</span>
  </div>

  <div class="week" id="week"></div>

  <div class="foot">O progresso fica salvo automaticamente neste dispositivo.</div>
</div>

<script>
const CATS = {
  flex: { color: 'var(--indigo)' },
  auto: { color: 'var(--clay)' },
  conc: { color: 'var(--teal)' }
};

const DAILY_ITEMS = [
  { id: 'leitura-matinal', cat: 'conc', text: 'Leitura matinal de 15 min, sem distrações' },
  { id: 'escrita-livre', cat: 'flex', text: 'Escrita livre por alguns minutos, sem se preocupar com estrutura' }
];

const DAY_SPECIFIC = {
  0: { id: 'diario', cat: 'auto', text: 'Registrar no diário uma frustração e uma conquista do dia' },
  1: { id: 'podcast', cat: 'conc', text: 'Ouvir um podcast por 15 min, sem interrupções' },
  2: { id: 'leitura-variada', cat: 'flex', text: 'Ler algo de um gênero ou autor diferente do habitual' },
  3: { id: 'documentario', cat: 'conc', text: 'Assistir a um trecho de documentário e anotar os pontos-chave' },
  4: { id: 'avaliar-projeto', cat: 'auto', text: 'Avaliar um projeto: o que repetiria e o que faria diferente' },
  5: { id: 'atividade-fisica', cat: 'auto', text: 'Praticar yoga, pilates ou outra atividade física com atenção plena' },
  6: { id: 'licoes-filme', cat: 'auto', text: 'Listar aprendizados de um filme assistido recentemente' }
};

const DAY_NAMES = ['Seg','Ter','Qua','Qui','Sex','Sáb','Dom'];

function isoDate(d){
  return d.toISOString().slice(0,10);
}

function getWeekDates(){
  const today = new Date();
  const dow = (today.getDay() + 6) % 7;
  const monday = new Date(today);
  monday.setDate(today.getDate() - dow);
  const dates = [];
  for(let i=0;i<7;i++){
    const d = new Date(monday);
    d.setDate(monday.getDate() + i);
    dates.push(d);
  }
  return dates;
}

const weekDates = getWeekDates();
const todayIso = isoDate(new Date());
const state = {};
let totalItems = 0;
let checkedCount = 0;

function updateRing(){
  const pct = totalItems ? Math.round((checkedCount/totalItems)*100) : 0;
  const circumference = 251.2;
  const offset = circumference - (pct/100)*circumference;
  document.getElementById('ring-progress').setAttribute('stroke-dashoffset', offset);
  document.getElementById('ring-pct').textContent = pct + '%';
}

async function toggleItem(key, checkboxEl, itemEl){
  const newVal = checkboxEl.checked;
  state[key] = newVal;
  checkedCount += newVal ? 1 : -1;
  itemEl.classList.toggle('checked', newVal);
  updateRing();
  try{
    await window.storage.set(key, newVal ? 'true' : 'false', false);
  }catch(e){
    console.error('Falha ao salvar', e);
  }
}

async function loadAndRender(){
  const weekEl = document.getElementById('week');
  weekEl.innerHTML = '';

  const dateFmt = new Intl.DateTimeFormat('pt-BR', { day: '2-digit', month: '2-digit' });

  for(let i=0;i<7;i++){
    const date = weekDates[i];
    const iso = isoDate(date);
    const items = [...DAILY_ITEMS, DAY_SPECIFIC[i]];
    totalItems += items.length;

    const dayEl = document.createElement('div');
    dayEl.className = 'day' + (iso === todayIso ? ' today' : '');

    const head = document.createElement('div');
    head.className = 'day-head';
    head.innerHTML = '<span class="dname">' + DAY_NAMES[i] + '</span><span class="ddate">' + dateFmt.format(date) + '</span>';
    dayEl.appendChild(head);

    for(const item of items){
      const key = 'routine:' + iso + ':' + item.id;
      let checked = false;
      try{
        const res = await window.storage.get(key, false);
        checked = res && res.value === 'true';
      }catch(e){
        checked = false;
      }
      state[key] = checked;
      if(checked) checkedCount++;

      const itemEl = document.createElement('div');
      itemEl.className = 'item' + (checked ? ' checked' : '');

      const checkbox = document.createElement('input');
      checkbox.type = 'checkbox';
      checkbox.checked = checked;

      const catDot = document.createElement('span');
      catDot.className = 'cat-dot';
      catDot.style.background = CATS[item.cat].color;

      const label = document.createElement('label');
      label.textContent = item.text;

      checkbox.addEventListener('change', () => toggleItem(key, checkbox, itemEl));
      itemEl.addEventListener('click', (e) => {
        if(e.target !== checkbox){
          checkbox.checked = !checkbox.checked;
          checkbox.dispatchEvent(new Event('change'));
        }
      });

      itemEl.appendChild(checkbox);
      itemEl.appendChild(catDot);
      itemEl.appendChild(label);
      dayEl.appendChild(itemEl);
    }

    weekEl.appendChild(dayEl);
  }

  updateRing();
}

loadAndRender();
</script>
</body>
</html>
'''
)
```
